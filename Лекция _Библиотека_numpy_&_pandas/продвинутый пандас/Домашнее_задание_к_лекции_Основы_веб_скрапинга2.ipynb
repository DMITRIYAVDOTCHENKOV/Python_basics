```python
def get_habr_posts():
    url = 'https://habr.com/ru/search/?q=%D0%BF%D0%B0%D0%B9%D1%82%D0%BE%D0%BD&target_type=posts&order=relevance'
    req = requests.get(url).text
    soup = BeautifulSoup(req)

    news = soup.find_all('div', class_='tm-article-snippet tm-article-snippet')

    habr_posts = pd.DataFrame()

    for el in news:
        title = el.find('a', 'tm-title__link').text
        link = el.find('a', 'tm-title__link').get('href')
        date = el.find('span', 'tm-article-datetime-published').text
        category = el.find('span').text
        row = {'date': date, 'title': title, 'link': link, 'category':category}
        habr_posts = pd.concat([habr_posts, pd.DataFrame([row])])
    return habr_posts.reset_index(drop=True)

get_habr_posts()
```

```python


import requests
from bs4 import BeautifulSoup
import time
import pandas as pd

def search_habr(query_list, num_pages):
    habr_posts = pd.DataFrame()
    URL = 'https://habr.com/ru/search/'
    for q in query_list:
        for page in range(1, num_pages + 1):
            params = {
                'q' : q,
                'page': f'page{page}'
            }
            req = requests.get(URL, params=params)
            time.sleep(0.3)
            soup = BeautifulSoup(req.text)
            articles = soup.find_all('article',class_='tm-articles-list__item')
            for article in articles:
                title = article.find('h2', class_='tm-megapost-sniper__title').text
                link ='https://habr.com' + article.find('a', class_='tm-megapost-sniper__link').get('href')
                date = article.find('span', class_='tm-article-datetime-published').text
                row = {'date': date, 'title': title, 'link': link}
                habr_posts = pd.concat([habr_posts, pd.DataFrame([row])])

    habr_posts = habr_posts.drop_duplicates('link').reset_index(drop=True)
    return habr_posts

search_habr(['python', 'data science'], num_pages=1)

```
